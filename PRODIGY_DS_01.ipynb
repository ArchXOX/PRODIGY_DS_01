```python
#Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Data
df=pd.read_excel('/content/drive/MyDrive/TASK 1/StudentPerformanceFactors.xlsm')
df.head()
```

```python
#Parents education level
sns.countplot(x='Parental_Education_Level',data=df)
plt.title('Bar plot of Parental Education Level')
plt.ylabel('Frequency')
plt.show()
```

From the above bar chart, it can be seen that there are a high number of parents who have completed only high school, followed by a fewer number of college graduates and even fewer parents possessing a postgraduate degree.
It can be inferred that the given categorical variable follows right skewed distribution for the data, so the data is asymmetrical.
